```python
zot_values_dict = {}
zot_values_dict["university"] = "institution"
zot_values_dict["place"] = "city"
zot_values_dict["date"] = "year"
zot_values_dict["accessDate"] : "created"
zot_values_dict["url"] = "websites"
zot_values_dict["bookTitle"] = "source"
zot_values_dict["publicationTitle"] = "source"

zot_items = []
for mend_item in types_examples:
    zot_type = mend_to_zot_types[mend_item["type"]]
    zot_type_template = zot_templates_dict[zot_type]
    # try to fill in the form automatically:
    extra = []
    for zot_key in zot_type_template.keys():
        try: 
            zot_type_template[zot_key] = mend_item[zot_key]
        except:
            pass
    try: zot_type_template["abtractNote"] = mend_item["abstract"]
    except: pass
    zot_creators = [] # dict: {"creatorType": "author", "firstName" : "lastName"}
    try:
        for author in mend_item["authors"]:
            zot_creators.append({"creatorType": "author", "firstName" : author["first_name"], "lastName" : author["last_name"]})
    except:
        pass
    try:
        for editor in mend_item["editors"]:
            zot_creators.append({"creatorType": "editor", "firstName" : author["first_name"], "lastName" : author["last_name"]})
    except:
        pass
    zot_type_template["creators"] = zot_creators
    try: zot_type_template["ISBN"] = mend_item["identifiers"]["isbn"]
    except: pass
    try: zot_type_template["ISSN"] = mend_item["identifiers"]["issn"]
    except: pass
    try: zot_type_template["DOI"] = mend_item["identifiers"]["doi"]
    except: pass
    try: zot_type_template["tags"] = [[{ "tag" : tag, "type" : "keywords"} for tag in mend_item["keywords"]] + [{ "tag" : tag, "type" : "tags"} for tag in mend_item["tags"]]]
    except: pass
    for key in zot_values_dict.keys():
        try:
            zot_type_template[key] = mend_item[zot_values_dict[key]]
        except:
            pass
    zot_items.append(zot_type_template)
    
# TO ADD:
# - accessDate (i.e. creationDate?)

# what to do with collections?

# where to put files?
```
