```python
!pip install beautifulsoup4
!pip install sddk
import sddk
```

```python
conf = sddk.configure()
```

```python
documents_dict = sddk.read_file("mendeley-mining/documents_dict_annots.json", "dict", conf)
```

```python
no_path_docs = []
folders_meshed = []
for doc in documents_dict.keys():
    try:
        folders_meshed.extend(documents_dict[doc]["paths"])    
    except:
        no_path_docs.append(doc)
```

```python
# to get only unique
folders_meshed = list(set(folders_meshed))
```

```python
path_lists = [folder.split("/") for folder in folders_meshed]
path_lists[:5]
```

```python
path_lists = sorted(path_lists)
```

```python
col_lev1 = list(set([path_list[0] for path_list in path_lists]))
col_lev1
```

```python
max([len(path) for path in path_lists])
```

```python
col_lev1_dict = [{"name" : col} for col in col_lev1]
zot.create_collections(col_lev1_dict)
```

```python
zot.create_collections([{"name": "test_collection"}])["successful"]["0"]["library"]["id"]
```

```python
main_cols = {}
for col in zot.collections():
    main_cols[col["data"]["name"]] = col["library"]["id"] 
main_cols
```

```python
for path in path_lists:
    for path_index in range(len(path)):
        try:
            if path_index == 0:
                last = zot.create_collections([{"name" : path[0]}])["successful"]["0"]["library"]["id"]
            else:
                last = zot.create_collections([{"name" : path[path_index], "parentCollection" : last}])["successful"]["0"]["library"]["id"]
        except:
            pass
```
